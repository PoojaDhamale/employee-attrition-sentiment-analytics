## Exploratory Data Analysis (EDA)

### 1. Understanding the Target Variable

Before analyzing other features, it is important to examine the distribution of the target variable (Attrition) to understand how many employees stayed and how many left the organization.

```python
df['Attrition'].value_counts()
```

```python
df['Attrition'].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Attrition', data=df)
plt.title('Distribution of Employee Attrition')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.show()
```

### Observation
- The majority of employees belong to the 'No' class.
- A relatively smaller proportion of employees have left the organization.
- The dataset is imbalanced; therefore, evaluation metrics such as Precision, Recall, F1-score, and ROC-AUC should be considered instead of relying solely on Accuracy.

### 5. Separating Numerical and Categorical Features

For further analysis, features are separated into numerical and categorical groups.

```python
num_cols=df.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols=df.select_dtypes(include='object').columns.tolist()
print(f"Numerical Features:{num_cols}\n")
print(f"Categorical Features:{cat_cols} ")
```

### 6. Univariate Analysis of Numerical Features

Univariate analysis helps understand the distribution, spread, skewness, and potential outliers in numerical variables.

```python
for col in num_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(7,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col.replace('_', ' '))
    plt.show()
```

### 7. Univariate Analysis of Categorical Features

Categorical analysis helps understand the frequency distribution of categories within each feature.

```python
for col in cat_cols:
    plt.figure(figsize=(8,5))
    sns.countplot(y=col,
                  data=df,
                  order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.show()
```

### 8. Numerical Features with Respect to Attrition

The objective is to determine whether the distribution of numerical variables differs between employees who stayed and those who left.

```python
for col in num_cols:
    plt.figure(figsize=(7,5))
    sns.boxplot(x='Attrition',
                y=col,
                data=df)
    plt.title(f'{col} vs Attrition')
    plt.show()
```

### 9. Categorical Features with Respect to Attrition

This analysis helps identify categories that have relatively higher employee attrition.

```python
for col in cat_cols:
    if col != 'Attrition':
        plt.figure(figsize=(10,5))
        sns.countplot(x=col,
                      hue='Attrition',
                      data=df)
        plt.xticks(rotation=45)
        plt.title(f'{col} vs Attrition')
        plt.show()
```

### 10. Correlation Analysis

Correlation analysis helps identify relationships among numerical variables and potential multicollinearity.

```python
plt.figure(figsize=(18,12))
sns.heatmap(df[num_cols].corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

### 11.Attrition Percentage by Categories

```python
important_cat_cols = [
    'OverTime',
    'Department',
    'JobRole',
    'BusinessTravel',
    'MaritalStatus',
    'Gender',
    'EducationField'
]
```

```python
for col in important_cat_cols:
    attr_pct = (
        pd.crosstab(
            df[col],
            df['Attrition'],
            normalize='index'
        ) * 100
    )

    attr_pct['Yes'].sort_values().plot(
        kind='barh',
        figsize=(8, 5)
    )

    plt.title(f'Attrition Percentage by {col}')
    plt.xlabel('Attrition Percentage (%)')
    plt.ylabel(col)
    plt.show()
```

## Summary of Findings

**Target variable.** The dataset is imbalanced — roughly 84% of employees stayed and 16% left. This imbalance means Accuracy alone is a misleading metric for any downstream model; Precision, Recall, F1-score, and ROC-AUC are more appropriate and are the metrics used in the modelling notebook.

**Data quality.** No missing values and no duplicate rows were found across the 1,470 employee records, so no imputation or deduplication was required.

**Numerical features.** Several features show outliers or right-skew on inspection (e.g. `MonthlyIncome`, `YearsAtCompany`, `YearsSinceLastPromotion`) consistent with typical compensation and tenure distributions, where a small group of long-tenured or highly-paid employees pulls the distribution's tail. `EmployeeCount`, `StandardHours`, and `EmployeeNumber` carry no useful signal (constant or pure identifier) and are dropped in the modelling notebook.

**Attrition drivers — numerical.** Boxplots split by `Attrition` show that employees who left tend to have lower `MonthlyIncome`, fewer `TotalWorkingYears`, and shorter `YearsAtCompany` than those who stayed — consistent with attrition being concentrated among newer, lower-earning employees.

**Attrition drivers — categorical.** The attrition-percentage-by-category charts (Section 11) show the clearest signal in this notebook:
- Employees who work `OverTime` leave at a substantially higher rate than those who don't.
- `Sales` shows a higher attrition rate than `Research & Development` or `Human Resources`.
- `Travel_Frequently` employees leave more often than those who rarely or never travel.
- Single employees show higher attrition than married or divorced employees.

These four categorical patterns directly match the top features identified by the Logistic Regression model in `attrition_modelling.ipynb` (OverTime, BusinessTravel, JobRole, MaritalStatus), which is a useful cross-check — the model is learning patterns that are visibly present in the raw data, not artifacts of preprocessing.

**Correlation structure.** The correlation heatmap shows expected positive relationships between tenure-related features (`YearsAtCompany`, `YearsInCurrentRole`, `YearsWithCurrManager`, `TotalWorkingYears`) and between `Age` and `TotalWorkingYears`. This is consistent with the multicollinearity flagged via VIF in the modelling notebook, where these same features were found to have VIF > 10 and were treated with caution when interpreting Logistic Regression coefficients.

**How this informs modelling.** This EDA motivates several modelling decisions made later in the project: engineering a single `Total_Satisfaction` score from five correlated satisfaction columns, checking VIF before trusting Logistic Regression coefficients, and treating `OverTime`, `BusinessTravel`, and `JobRole` as priority features given their visibly large attrition gaps in this notebook.
